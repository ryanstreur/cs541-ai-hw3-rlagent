```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("episodes.csv")
fig, ax = plt.subplots()
ax.scatter(df[["episode_id"]], df[["episode_reward"]])
ax.plot(df[["episode_id"]], df[["running_avg"]], color="orange")


ax.set_xlabel("Episode ID")
ax.set_ylabel("Total Episode Reward")
```
