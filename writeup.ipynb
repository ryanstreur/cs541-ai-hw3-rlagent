# Q-Learning Experiment

This project is an attempt to implement a Q-Learning system for a grid-world simulation of a robot moving around an *n* by *n* grid where each space in the grid has a probability of 0.5 to contain a Can. 

To the greatest extent possible this program was written to conform to the specifications in the assignment. The robot has five sensors:

- Current
- North
- South
- East
- West

Each space on the grid has two possible states:

- Empty
- Can

If a robot attempts to sense a space outside the bounds of the grid, that space is marked as "Wall."

The robot can perform five actions:

- MoveNorth
- MoveSouth
- MoveEast
- MoveWest
- PickUp

The rewards are as follows:

| Action | Reward |
|:--|:--|
| Robot performs PickUp while Current is Can | 10 |
| Robot moves towards a "Wall" | -5 |
| Robot performs PickUp While Current is not Can | -1 |

The remaining parameters of the experiment should match the parameters in the assignment. Below are the plots and summary statistics for the episode score sums as described in the assignment.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("episodes.csv")
test_fig, test_ax = plt.subplots()
test_ax.set_title("Training Reward Plot")
test_ax.scatter(df[["episode_id"]], df[["episode_reward"]], label="Episode Score")
test_ax.plot(df[["episode_id"]], df[["running_avg"]], color="orange", label="Avg of Last 100 Episodes")
test_ax.plot(df[["episode_id"]], df[["crash_count"]], color="red", label="Number of Crashes")


test_ax.set_xlabel("Episode ID")
test_ax.set_ylabel("Total Episode Reward")
test_ax.legend()
```
