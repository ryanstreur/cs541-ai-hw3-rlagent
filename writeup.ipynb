```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("episodes.csv")
fig, ax = plt.subplots()
ax.scatter(df[["episode_id"]], df[["episode_reward"]], label="Episode Score")
ax.plot(df[["episode_id"]], df[["running_avg"]], color="orange", label="Avg of Last 100 Episodes")
ax.plot(df[["episode_id"]], df[["crash_count"]], color="red", label="Number of Crashes")


ax.set_xlabel("Episode ID")
ax.set_ylabel("Total Episode Reward")
ax.legend()
```
